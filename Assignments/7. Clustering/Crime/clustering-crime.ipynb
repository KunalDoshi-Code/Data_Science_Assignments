```python
# Normalize heterogenous numerical data using standard scalar fit transform to dataset
df_crime_norm=StandardScaler().fit_transform(df_crime)
df_crime_norm
```

```python
# DBSCAN Clustering
dbscan=DBSCAN(eps=1,min_samples=4)
dbscan.fit(df_crime_norm)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
# Adding clusters to dataset
df_crime['clusters']=dbscan.labels_
df_crime
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(df_crime['clusters'],df_crime['UrbanPop'], c=dbscan.labels_) 
```

#### **crew plot/elbow graph**

```python
k=list(range(2,15))
k
TWSS=[]
for i in k:
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df_crime_norm)
    WSS=[]
    for j in range(i):
        WSS.append(sum(cdist(df_crime_norm[kmeans.labels_==j], kmeans.cluster_centers_[j].reshape(1, df_crime_norm.shape[1]), metric='euclidean')))
    TWSS.append(sum(WSS))
plt.figure(figsize=(16, 9))
plt.plot(k,TWSS,'ro-');plt.xlabel('k values');plt.ylabel('TWSS')
```

### **k-means clustering**

```python
model = KMeans(n_clusters=6).fit(df_crime_norm)
cluster_labels = pd.DataFrame(model.labels_, columns=['cluster'])
final_data = pd.concat([cluster_labels, df_crime], axis=1)

insights=final_data.groupby(final_data.clusters).mean()
insights
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
dendrogram = sch.dendrogram(sch.linkage(df_crime_norm, method='ward'))
```

```python
hc1 = AgglomerativeClustering(n_clusters=4, affinity = 'euclidean', linkage = 'ward')
```
